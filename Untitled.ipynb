# Feature Selection for Customer Segmentation

Before applying clustering, it is important to retain only features that provide meaningful information about customer behavior. Irrelevant columns are removed to improve model interpretability and ensure that clustering reflects real-world patterns.

**Dropped Columns:**

* **Customer Name:** A textual identifier with no analytical value for clustering or pattern recognition.
* **Churn:** A target variable used in supervised learning. Since clustering is an unsupervised technique, including it would introduce bias and prevent the model from discovering natural groupings.

**Retained Columns:**

* **Customer ID:** Retained solely for grouping transactions belonging to the same customer during feature engineering. It is not used as an input feature for clustering.
* **Purchase Date:** Used to derive recency, indicating how recently a customer made a purchase.
* **Product Category:** Provides insight into customer preferences.
* **Product Price, Quantity, Total Purchase Amount:** Capture purchasing behavior and spending patterns.
* **Payment Method:** Reflects transaction behavior.
* **Age and Gender:** Represent customer demographics.
* **Returns:** Indicates customer satisfaction and post-purchase behavior.

By retaining only relevant behavioral and demographic features while preserving identifiers for grouping, the segmentation process becomes both meaningful and aligned with real-world business objectives.


````python
df["Returns"] = df["Returns"].fillna(0)
````

### Handling Missing Values in the `Returns` Column

The `Returns` column represents the number of items a customer has returned. Missing values in this column are interpreted as no recorded returns rather than unknown data.

Instead of removing these rows or imputing arbitrary values, missing entries were filled with 0. This approach is based on business logic, where the absence of a return record reasonably indicates that the customer did not return any items.

Filling missing values with 0 ensures data consistency, preserves all customer records, and maintains the integrity of behavioral analysis for segmentation.


````python
# Define snapshot date
snapshot_date = df["Purchase Date"].max() + pd.Timedelta(days=1)

# Create RFM
rfm = df.groupby("Customer ID").agg({
    "Purchase Date": lambda x: (snapshot_date - x.max()).days,
    "Customer ID": "count",
    "Total Purchase Amount": "sum"
})

# Rename columns
rfm.columns = ["Recency", "Frequency", "Monetary"]
````

### RFM Feature Engineering (Customer Behavior Modeling)

To enable meaningful customer segmentation, transaction-level data was transformed into customer-level behavioral features using the RFM framework.

The dataset initially contains multiple rows per customer, each representing a transaction. However, clustering requires a single representation per customer. RFM achieves this by aggregating transaction data into three key metrics:

* **Recency (R):** How recently a customer made a purchase
* **Frequency (F):** How often a customer makes purchases
* **Monetary (M):** How much a customer spends

This transformation converts raw transactional data into structured behavioral profiles, making it easier to identify patterns and group similar customers.

#### Implementation Breakdown

```python
# Define snapshot date
snapshot_date = df["Purchase Date"].max() + pd.Timedelta(days=1)
```

A reference date is created by taking the latest purchase date in the dataset and adding one day. This ensures that the most recent customer has a recency value of at least 1.

```python

```

The dataset is grouped by Customer ID so that all transactions belonging to a single customer are aggregated:

* Recency is calculated as the number of days since the customer's most recent purchase
* Frequency is computed as the total number of transactions per customer
* Monetary is calculated as the total amount spent by the customer

```python

```

The resulting columns are renamed for clarity and standardization.

This process transforms raw transaction data into meaningful customer-level features, which are essential for effective clustering and segmentation.

### Understanding RFM Distribution and Handling Skewness

After constructing the RFM (Recency, Frequency, Monetary) features, exploratory analysis was performed to understand their distributions.

The summary statistics and histograms revealed that all three features are **left-skewed**, meaning that most customers have relatively low values, while a small number of customers exhibit very high values (e.g., high spending or frequent purchases).

This skewness is common in real-world transactional data, where a minority of customers contribute disproportionately to revenue.

However, skewed data can negatively impact clustering algorithms such as K-Means, which rely on distance calculations. Features with large ranges or extreme values can dominate the clustering process, leading to biased or less meaningful groupings.

To address this, a **log transformation** was applied to the RFM features. This transformation compresses large values while preserving the relative differences between observations, resulting in a more balanced distribution.

By reducing skewness, log transformation improves the effectiveness of clustering and ensures that all features contribute more equally to the segmentation process.

After transformation, feature scaling is applied to standardize the data, ensuring that differences in magnitude across features do not distort distance calculations.


````python
rfm_log.hist(figsize=(10,5))
````

After Applying Log Transformation, it makes the data less skewed and more usable

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_log)
````

### Now we move to clustering

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

plt.figure()
plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
````

### Determining Optimal Number of Clusters (Elbow Method)

To determine the optimal number of clusters for customer segmentation, the Elbow Method was applied using the K-Means clustering algorithm.

K-Means requires specifying the number of clusters (K) in advance. To identify the most suitable value of K, the model was trained with different cluster sizes ranging from 1 to 10. For each value of K, the **inertia** (within-cluster sum of squares) was calculated, which measures how tightly data points are grouped within each cluster.

As the number of clusters increases, inertia decreases because data points are assigned to closer cluster centers. However, beyond a certain point, the rate of improvement slows down significantly.

This behavior is visualized by plotting the number of clusters against inertia. The optimal value of K is identified at the "elbow point" — where the curve begins to flatten, indicating diminishing returns from adding more clusters.

From the plot, a clear bend was observed at **K = 4**, suggesting that four clusters provide a good balance between model simplicity and clustering performance.

Therefore, **K = 4** was selected as the optimal number of clusters for segmenting customers.


````python
kmeans = KMeans(n_clusters=4, random_state=42)

rfm["Cluster"] = kmeans.fit_predict(rfm_scaled)
````

````python
rfm.groupby("Cluster").mean()
````

### Customer Segmentation Insights

After applying K-Means clustering with K = 4, customers were grouped into distinct segments based on their Recency, Frequency, and Monetary (RFM) values.
Each cluster represents a unique behavioral pattern:
Cluster 3 – High-Value Active Customers:
These customers have purchased recently, buy frequently, and spend significantly. They are the most valuable segment and should be prioritized for retention and loyalty programs.
Cluster 0 – Loyal but At-Risk Customers:
These customers have high purchase frequency and spending but have not made recent purchases. They represent a valuable segment that may require re-engagement strategies.
Cluster 2 – Moderate Customers:
These customers exhibit average purchasing behavior across all metrics. They present opportunities for upselling and engagement improvement.
Cluster 1 – Low-Value / At-Risk Customers:
These customers have low frequency, low spending, and have not purchased recently. They may require targeted marketing or may be deprioritized depending on business strategy.
This segmentation provides actionable insights for personalized marketing, customer retention, and revenue optimization.

### Applying PCA to help with visualization

````python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

rfm_pca = pca.fit_transform(rfm_scaled)
````

````python
### Create DataFrame for plotting
rfm_pca_df = pd.DataFrame(rfm_pca, columns=["PC1", "PC2"])

rfm_pca_df["Cluster"] = rfm["Cluster"].values
````

````python
### Plutting Clusters
plt.figure(figsize=(8,6))

plt.scatter(
    rfm_pca_df["PC1"],
    rfm_pca_df["PC2"],
    c=rfm_pca_df["Cluster"],
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Customer Segmentation (K-Means Clusters)")
plt.show()
````

### Cluster Visualization using PCA

To visualize customer segments, Principal Component Analysis (PCA) was applied to reduce the dimensionality of the scaled RFM features from three dimensions to two.

This transformation preserves the most important variance in the data while enabling effective visualization.

The resulting 2D plot shows the distribution of customers across different clusters, providing a clear visual representation of the segmentation structure. Although some overlap may exist, distinct groupings can be observed, validating the clustering results.
